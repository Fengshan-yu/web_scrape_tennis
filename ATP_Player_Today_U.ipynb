```python
import re
percentage = r"\d{2}\.\d"
number2 = r"\d+(,\d{3})*\.*\d*"

def get_percentage(record):
    if record == "N/A":
        pass
    elif record == None:
        pass
    else:
        return float(re.search(percentage, record).group())
    
def get_prize(money):
    if money == "N/A":
        pass
    else:
        return int(re.search(number2, money).group().replace(",", ""))
    
def get_birthday(phrase):
    if phrase == "N/A":
        pass
    else:
        return phrase.lstrip("( ").split(" ")[0]
    
def get_birthplace(phrase):    
    if phrase == "N/A":
        pass
    else:
        return " ".join(phrase.split(" ")[8:])
    
for i, player in enumerate(player_info_list):
    if i<=20:
        player["prize"]=get_prize(player.get("Prize money", "N/A")
        player["Win Percentage"]=get_percentage(player.get("Career record", "N/A"))
        player["Birthday"]=get_birthday(player.get("Born", "N/A"))
        player["Birth Place"]=get_birthplace(player.get("Born", "N/A"))
```
